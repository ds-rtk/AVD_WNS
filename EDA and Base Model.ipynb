```python
# Function to calculate missing values by column# Funct 

def missing_value_table(df):
    mis_value = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_table = pd.concat([mis_value, mis_val_percent], axis = 1)
    mis_ren_clmn = mis_val_table.rename(columns = {0 : 'Missing Value', 1 : "percent missing"})
    mis_ren_clmn = mis_ren_clmn[mis_ren_clmn.iloc[:,1] !=0].sort_values('percent missing',ascending = False).round(1)
    return mis_ren_clmn
```

```python
# Number of CV folds (to generate level-one data for stacking)
nfolds = 10

# There are a few ways to assemble a list of models to stack together:
# 1. Train individual models and put them in a list
# 2. Train a grid of models
# 3. Train several grids of models
# Note: All base models must have the same cross-validation folds and
# the cross-validated predicted values must be kept.


# 1. Generate a 2-model ensemble (GBM + RF)

# Train and cross-validate a GBM
my_gbm = H2OGradientBoostingEstimator(distribution="bernoulli",
                                      ntrees=100,
                                      max_depth=3,
                                      min_rows=2,
                                      learn_rate=0.001,
                                      nfolds=nfolds,
                                      fold_assignment="Modulo",
                                      keep_cross_validation_predictions=True,
                                      seed=1)
my_gbm.train(x=x, y=y, training_frame=train_raw)


# Train and cross-validate a RF
my_rf = H2ORandomForestEstimator(ntrees=100,
                                 nfolds=nfolds,
                                 fold_assignment="Modulo",
                                 keep_cross_validation_predictions=True,
                                 seed=1)
my_rf.train(x=x, y=y, training_frame=train_raw)


# Train a stacked ensemble using the GBM and GLM above
ensemble = H2OStackedEnsembleEstimator(model_id="ensemble_baseline",
                                       base_models=[my_gbm, my_rf,best_gbm1,aml.leader])
ensemble.train(x=x, y=y, training_frame=train_raw)
```
